```python
df = pd.read_excel('transactionData.xlsx', sheetname='Sheet1')
#type(df.iloc[i]["Transaction Amount"])
df
```

```python
ds={}
ds1={}
dl={}
l = ["A","B","C"]
for i in l:
    ds[i]={}
    ds1[i]={}
    k = ["Travel","Dining","Fuel","Online Shopping","Shopping"]
    for j in k:
        ds[i][j]=0 
        ds1[i][j]=0
    ds[i]["Due"]=0
    ds[i]["Pay"]=0
```

```python
for i in range(len(df)):
    if(int(df.iloc[i]["Customer ID"])>11010):
        dl[df.iloc[i]["Customer ID"]]=df.iloc[i]["Age"]
        if(df.iloc[i]["Age"] >= 15 and df.iloc[i]["Age"]<20):
            p="C"
            c=df.iloc[i]["Category"]
            l=ds[p][c]
            l += int(df.iloc[i]["Transaction Amount"])
            ds[p][c] = l
        elif(df.iloc[i]["Age"] >= 20 and df.iloc[i]["Age"]<=39):
            p="B"
            c=df.iloc[i]["Category"]
            ds[p][c] += int(df.iloc[i]["Transaction Amount"])
        elif(df.iloc[i]["Age"] >= 40 and df.iloc[i]["Age"]<=55):
            p="A"
            c=df.iloc[i]["Category"]
            ds[p][c] += int(df.iloc[i]["Transaction Amount"])
```
